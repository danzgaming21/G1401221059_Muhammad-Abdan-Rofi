```python
import pandas as pd
df = pd.read_excel('/content/drive/My Drive/Colab Notebooks/upah.xlsx')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

df['Rata_rata_upah'] = df[['upah2015', 'upah2016', 'upah2017', 'upah2018', 'upah2019', 'upah2020', 'upah2021', 'upah2022']].mean(axis=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Urutkan DataFrame berdasarkan 'Rata_rata_upah' secara menaik
sorted_df = df.sort_values(by='Rata_rata_upah', ascending=True)

# Tetapkan skala warna
colors = plt.cm.Blues(np.linspace(0.2, 1, len(df)))

plt.figure(figsize=(10, 6))
plt.barh(sorted_df['provinsi'], sorted_df['Rata_rata_upah'], color=colors)
plt.xlabel('Rata-rata Jumlah Upah per jam')
plt.ylabel('Provinsi')
plt.title('Rata-rata Jumlah Upah per jam setiap Provinsi dari tahun 2015-2022')
plt.tight_layout()
plt.show()
```

barchart diatas menunjukan bahwa rataan upah per jam dari tahun 2015-2022 paling tinggi di daerah DKI Jakarta dan palin rendah daerah Jawa tengah. Hal tersebut memberikan petunjuk kepada kita untuk menentukan rencana tempat kerja yang upah kerja nya paling tinggi di Indonesia.

```python
import matplotlib.pyplot as plt

df['data2022'] = df['upah2022']  # Tidak perlu menggunakan double brackets di sini
plt.figure(figsize=(12, 6))
plt.plot(df['provinsi'], df['upah2022'], marker='o', linestyle='-')  # Menggunakan df langsung
plt.xlabel('Provinsi')
plt.ylabel('Upah per jam tahun 2022')
plt.title('Upah per jam tahun 2022 per Provinsi')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Linechart diatas menunjukan jumlah upah per jam tahun 2022 setiap provinsi, bisa kita lihat bahwa yang paling tinggi di daerah DKI Jakarta berjumlah Rp 32.685 per jam.
